```python
import pandas as pd
DF= pd.read_csv("Netflix_final.csv", parse_dates=['date_added'], dayfirst=True)
DF.dtypes
```

## 📈 Netflix Content Growth
- Few shows before 2010  
- Rapid increase after 2015  
- Slight dip in 2020  


```python
df['release_year'].value_counts().sort_index().plot(kind='line')
```

```python
movies=DF[DF['type']=='Movie']
Totalmovies=movies['date_added'].dt.year.value_counts().sort_index()
Totalmovies
#i again did this line to avoid any errors because of the delay of things.
```

```python
shows=DF[DF['type']=='TV Show']
Totalshows=shows['date_added'].dt.year.value_counts().sort_index()
Totalshows
```

## 🎬 Movies vs TV Shows (Yearly Trend)

- 📌 Movies are released **more frequently** than TV Shows every year.  
- 🎥 Movies dominate the content library, especially after **2010**.  
- 📺 TV Shows have grown steadily, but still fewer compared to Movies.  
- 🔎 Overall, the trend shows **Movies are the major focus** of Netflix content releases.


```python
import matplotlib.pyplot as plt
import numpy as np
all_years = sorted(set(Totalmovies.index).union(set(Totalshows.index)))

# Align both with the full year range (fill missing years with 0)
movies_aligned = Totalmovies.reindex(all_years, fill_value=0)
shows_aligned = Totalshows.reindex(all_years, fill_value=0)

# Plot side by side bars
x = np.arange(len(all_years))
bar_width = 0.4

plt.bar(x - bar_width/2, movies_aligned.values, width=bar_width, label="Movies", color="#FF6F61")
plt.bar(x + bar_width/2, shows_aligned.values, width=bar_width, label="TV Shows", color="#6B5B95")

plt.xticks(x, all_years, rotation=45)
plt.xlabel("Years")
plt.ylabel("Count")
plt.title("Movies vs TV Shows per Year on Netflix")
plt.legend()
plt.tight_layout()
plt.show()
#This is the code to compare movies and TV Shows released per year.
```

```python
ratio_per_year=Totalmovies/Totalshows.replace(0,float('nan'))
ratio_per_year #To display release ration between movies and TV Shows.
```

```python
# Group and count ratings by type
counts = DF.groupby(['type', 'rating']).size().unstack(fill_value=0)

# Plot without transpose: ratings on x-axis
counts.plot(kind='bar', figsize=(12,6))

# Labels & Title
plt.title("Ratings Distribution: Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

## 🎥 Directors with Most Content
- Director Rajiv Chilaka has the most Netflix titles.  
- Director Suhas Kadav mainly makes kids’ shows.  
- Most active directors are from 2010s onwards.


```python
#The assign method lets you create a new column on the fly.
#The .explode() function takes a list in one row and creates multiple rows.
Directors=(
    DF[DF['director']!='Not found']
    .assign(director=DF['director'].str.split(','))
    .explode('director')
)
Directors['director']=Directors['director'].str.strip()
print(Directors['director'].value_counts().head(20))
#This displays the top most Directors over the years.
```

```python
DF['title_len']=DF['title'].str.len()
longest_titles=(
    DF.sort_values('title_len',ascending=False)
    [['title','title_len','type','release_year','director']].head(10)
)
print(longest_titles.to_string(index=False))
#It will display the longest title 
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.barh(longest_titles['title'], longest_titles['title_len'])
plt.xlabel("Title Length")
plt.ylabel("Title")
plt.title("Top 10 Longest Netflix Titles")
plt.gca().invert_yaxis()  # longest at top
plt.show()
#i done this to visualize a little bit it confusing.
```

## 📊 NumPy Statistics on Netflix Dataset

- 🔢 **Descriptive stats** (mean, median, mode, min, max) help us understand the dataset better.  
- ⏳ **Release years**: show the spread of content across decades, highlighting Netflix’s recent growth.  
- ✍️ **Title lengths**: reveal how movie/show titles vary in size (from short to very long titles).  
- 🎯 These stats provide a **numerical summary** that complements the visual charts and tables.  


```python
title_lengths=DF['title_len'].values
print("Numpy stats of Title lengths:")
print("Mean length :",np.mean(title_lengths))
print("Median length:",np.median(title_lengths))
print("Max length:",np.max(title_lengths))
print("Min length:",np.min(title_lengths))
print("percentile:",np.percentile(title_lengths,90))
```
